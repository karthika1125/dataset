```python
import pandas as pd
df=pd.read_csv('Movies Dataset.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
column_to_drop=['credits', 'poster_path','backdrop_path','keywords','tagline','overview','runtime','recommendations']
```

```python
df_cleaned=df.drop(columns=column_to_drop)
df_cleaned
```

```python
df_cleaned.isnull().sum()
```

```python
numerical=df_cleaned.select_dtypes(include=['number']).columns
numerical
```

```python
categorical=df_cleaned.select_dtypes(include=['object']).columns
categorical
```

```python
df_cleaned[numerical]=df_cleaned[numerical].fillna(df_cleaned[numerical].mode)
```

```python
df_cleaned[categorical]=df_cleaned[categorical].fillna('unknown')
```

```python
df_cleaned.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label={}
for col in categorical:
    le=LabelEncoder()
    df_cleaned[col]=le.fit_transform(df_cleaned[col])
    label[col]=le
df_cleaned.head()    
```

```python
from sklearn.preprocessing import StandardScaler
scaler={}
for col in numerical:
    sc=StandardScaler()
    df_cleaned[col]=sc.fit_transform(df_cleaned[[col]])
    scaler[col]=sc
df_cleaned.head()    
```
