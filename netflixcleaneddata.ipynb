```python
import pandas as pd
df=pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
numerical_column=df.select_dtypes(include=['number']).columns
numerical_column
```

```python
categorical_column=df.select_dtypes(include=['object']).columns
```

```python
categorical_column
```

```python
df[categorical_column]=df[categorical_column].fillna('unknown')
```

```python
df[numerical_column]=df[numerical_column].fillna(df[numerical_column].median)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
from sklearn.preprocessing import LabelEncoder
label={}
for col in categorical_column:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
    label[col]=le
df    
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df['release_year']=scaler.fit_transform(df[['release_year']])
df
```
